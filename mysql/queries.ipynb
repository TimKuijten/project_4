```python
all_restaurants_languages = pd.read_sql_query("""
SELECT DISTINCT t.*, r.Cuisine_type as 'Cuisine', r.Rating as 'Total Rating'
FROM reviews t
LEFT JOIN restaurant_top r
ON t.Restaurant = r.name;
""", engine)
```

```python
import pandas as pd
import stylecloud
from IPython.display import Image
```

```python
high_arabic = pd.read_csv('../csv/mysql/high_arabic_reviews.csv')
high_chinese = pd.read_csv('../csv/mysql/high_chinese_reviews.csv')
high_dutch = pd.read_csv('../csv/mysql/high_Dutch_reviews.csv')
high_english = pd.read_csv('../csv/mysql/high_english_reviews.csv')
high_french = pd.read_csv('../csv/mysql/high_french_reviews.csv')
high_german = pd.read_csv('../csv/mysql/high_german_reviews.csv')
high_greek = pd.read_csv('../csv/mysql/high_greek_reviews.csv')
high_hebrew = pd.read_csv('../csv/mysql/high_hebrew_reviews.csv')
high_hindi = pd.read_csv('../csv/mysql/high_hindi_reviews.csv')
high_italian = pd.read_csv('../csv/mysql/high_italian_reviews.csv')
high_korean = pd.read_csv('../csv/mysql/high_korean_reviews.csv')
high_norwegian = pd.read_csv('../csv/mysql/high_norwegian_reviews.csv')
high_portuguese = pd.read_csv('../csv/mysql/high_portuguese_reviews.csv')
high_russian = pd.read_csv('../csv/mysql/high_russian_reviews.csv')
high_spanish = pd.read_csv('../csv/mysql/high_spanish_reviews.csv')
high_swedish = pd.read_csv('../csv/mysql/high_swedish_reviews.csv')
high_thai = pd.read_csv('../csv/mysql/high_thai_reviews.csv')
high_turkish = pd.read_csv('../csv/mysql/high_turkish_reviews.csv')
high_vietnamese = pd.read_csv('../csv/mysql/high_vietnamese_reviews.csv')
```

```python

high_asia = pd.concat([high_chinese, high_korean, high_thai, high_vietnamese])
high_europe = pd.concat([high_dutch, high_french, high_german, high_greek, high_italian, high_norwegian, 
                         high_portuguese, high_swedish])
```

```python
stylecloud.gen_stylecloud(text=high_asia["Translated Review"].str.cat(sep=' '),
                          icon_name='fas fa-thumbs-up',
                          max_words = 50,
                          palette='colorbrewer.sequential.BuGn_9',
                          background_color='black',
                          gradient='horizontal')
Image(filename='stylecloud.png')                          
```

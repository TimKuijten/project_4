```python
import pandas as pd
import stylecloud
from IPython.display import Image
```

```python
low_arabic = pd.read_csv('../csv/mysql/low_arabic_reviews.csv')
low_chinese = pd.read_csv('../csv/mysql/low_chinese_reviews.csv')
low_dutch = pd.read_csv('../csv/mysql/low_Dutch_reviews.csv')
low_english = pd.read_csv('../csv/mysql/low_english_reviews.csv')
low_french = pd.read_csv('../csv/mysql/low_french_reviews.csv')
low_german = pd.read_csv('../csv/mysql/low_german_reviews.csv')
low_greek = pd.read_csv('../csv/mysql/low_greek_reviews.csv')
low_hebrew = pd.read_csv('../csv/mysql/low_hebrew_reviews.csv')
low_hindi = pd.read_csv('../csv/mysql/low_hindi_reviews.csv')
low_italian = pd.read_csv('../csv/mysql/low_italian_reviews.csv')
low_korean = pd.read_csv('../csv/mysql/low_korean_reviews.csv')
low_norwegian = pd.read_csv('../csv/mysql/low_norwegian_reviews.csv')
low_portuguese = pd.read_csv('../csv/mysql/low_portuguese_reviews.csv')
low_russian = pd.read_csv('../csv/mysql/low_russian_reviews.csv')
low_spanish = pd.read_csv('../csv/mysql/low_spanish_reviews.csv')
low_swedish = pd.read_csv('../csv/mysql/low_swedish_reviews.csv')
low_thai = pd.read_csv('../csv/mysql/low_thai_reviews.csv')
low_turkish = pd.read_csv('../csv/mysql/low_turkish_reviews.csv')
low_vietnamese = pd.read_csv('../csv/mysql/low_vietnamese_reviews.csv')
```

```python

low_asia = pd.concat([low_chinese, low_korean, low_thai, low_vietnamese])
low_europe = pd.concat([low_dutch, low_french, low_german, low_greek, low_italian, low_norwegian, 
                         low_portuguese, low_swedish])
```

```python
stylecloud.gen_stylecloud(text=low_asia["Translated Review"].str.cat(sep=' '),
                          icon_name='fas fa-thumbs-down',
                          max_words = 50,
                          palette='colorbrewer.sequential.OrRd_9',
                          background_color='black',
                          gradient='horizontal')
Image(filename='stylecloud.png')                          
```
